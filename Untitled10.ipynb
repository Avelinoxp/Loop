```python
#ATIVIDADE1-Calculando bônus dos funcionários

vendas_funcionarios = [2800, 3200, 1500, 4000, 5000, 1000, 3500]
metas = 3000
percentual_bonus = 0.05
for venda in vendas_funcionarios:
    if venda >= metas:
     bonus = percentual_bonus * venda
    else:
        bonus = 0
    print(bonus)
```

```python
#2-Contagem de produtos vendidos acima da meta

diarias = [2000,3000,2500,4000,1800,5000,2700]
meta = float (2500)
d_acima_meta = 0
for diaria in diarias:
    if diaria > meta:
        d_acima_meta = d_acima_meta +1
    print(d_acima_meta)

```

```python
#3-Cálculo de comissão dos vendedores

venda_vendedores = [1500, 2500, 1800, 3200, 4000, 1200, 2800]
meta = float(0.08)
for venda in venda_vendedores:
    if venda > meta:
     bonus = venda * meta
    else:
     bonus = 0
    print(bonus)
```

```python
#4-Contando vendas que atingiram a meta

lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = (1200)
contador = 0
for venda in lista_vendas:
    if venda > meta:
     #contador = contador + 1
     contador += 1
print('Quantidade de dias que bateram a meta:', contador)

```

```python
#5-Aplicando um bônus extra para grandes vendas

lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = 1200
percentual_bonus = 0.1
bonificacao = 50

for venda in lista_vendas:
    bonus = 0
    if venda >= meta:
        bonus = venda * percentual_bonus
        if venda >= 2000:
            bonus += bonificacao
    print(f'Venda: {venda} -> Bonus : R${bonus}')

```
