```python
#ESTRUTURA DE LOOP WHILE
soma = 0
numero = 0
while numero != -1:
    numero = int(input('Digite um número (-1 para parar):'))
    if numero != -1:
        soma += numero
print('A soma dos números é:', soma)

#while numero != -1:
#Esse é o laço de repetição.
#Enquanto o número digitado não for -1, o loop continua rodando.
#if numero != -1:
#Antes de somar, o programa verifica se o número digitado não é -1.
#Isso evita que o -1 (que serve só para parar o programa) entre na soma.
#soma += numero
#Se o número for diferente de -1, ele é somado à variável soma
```

```python
#Atividade3
soma = 0
numero_digitado = 0  # Inicializa com um valor diferente de -5 para entrar no loop

while numero_digitado != -5:
    numero_digitado = int(input("Digite um número inteiro (-5 para parar): "))
    if numero_digitado != -5:
        soma += numero_digitado

print(f"A soma dos números digitados (excluindo -5) é: {soma}")
```
